```python
import pandas as pd
import numpy as np
```

```python
data =pd.read_csv("perceptron.data", names=["X1","X2","X3","X4","Y"])
```

```python
w0 = [0,0,0,0]
b=0
precision = 0
step_size = 1
w_list = []
b_list = []
loss=[]
l = len(data)-1
```

```python
def Loss(m,n,w0,b):
    temp=data[m:n]
    temp['calc'] = temp.apply(lambda x: 1 if(-1*x.values[4]*(np.dot(x.values[:4],np.array(w0))+b))>=0 else 0,axis =1)
    delta_w = temp.apply(lambda x: x.values[5]*x.values[4]*x.values[:4],axis=1)
    print("delta_w:", delta_w)
    delta_b = temp.apply(lambda x: x.values[5]*x.values[4],axis=1)
    print("delta_b:", delta_b)
    del temp
    return (sum(delta_w),sum(delta_b))
```

```python
def GradientDescent(m,n,w0,b,step_size):
    delta_w, delta_b = Loss(m,n,w0,b)
#     print (delta_w, delta_b)
    if all(delta_w==precision) and (delta_b==precision):
        return
    w0+=step_size*delta_w
#     print(w0)
    w_list.append(list(w0))
    b+=step_size*delta_b
#     print(b)
    b_list.append(b)
#     data['calc'][m:n] = data[m:n].apply(lambda x: 1 if(-1*x.values[4]*(np.dot(x.values[:4],np.array(w0))+b))>=0 else 0,axis =1)
    GradientDescent(0,len(data),w0,b,step_size)
    
```

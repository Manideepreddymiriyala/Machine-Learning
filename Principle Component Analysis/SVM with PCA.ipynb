```python
import numpy as np
import pandas as pd
from sklearn.svm import LinearSVC
```

```python
Train = pd.read_csv(r'C:\Users\miriy\Downloads\sonar_train.data', header = None)
Test  = pd.read_csv(r'C:\Users\miriy\Downloads\sonar_test.data', header = None)
Valid = pd.read_csv(r'C:\Users\miriy\Downloads\sonar_valid.data', header = None)
```

```python
n = len(Train)
m =len(Train.columns)
```

```python
X_train, y_train = np.array(Train.iloc[:,0:60]),np.array(Train.iloc[:,60])
y_train[y_train == 2] = -1
X_test, y_test = np.array(Train.iloc[:,0:60]),np.array(Train.iloc[:,60])
y_test[y_test==2]=-1
X_valid, y_valid = np.array(Train.iloc[:,0:60]),np.array(Train.iloc[:,60])
y_valid[y_valid==2]=-1
```

```python
def normalize_data(X, X_train):
    X_norm = X - X_train.mean(axis=0)
    X_norm = X_norm / X_train.std(axis=0)
    return X_norm

def feature_select(D, k, s, eig_vals, eig_vecs):
    """
    Select s features using the top k eigenvectors.
    eig_vals and eig_vecs must be sorted.
    Return an array of features indices.
    D: number of features in the training set.
    """
    idx = eig_vals.argsort()[::-1]
    eig_vals = eig_vals[idx]
    eig_vecs = eig_vecs[:,idx]
    top_vecs = eig_vecs[:k]
    feature_dist = np.sum(top_vecs**2, axis=0) / k
    # sample s features using np.random.choice
    features = np.random.choice(D, s, p=feature_dist)
    # get the unique feature indices
    features = np.unique(features)
    return features

```

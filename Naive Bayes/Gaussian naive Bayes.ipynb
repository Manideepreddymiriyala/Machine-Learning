```python
import numpy as np
import pandas as pd
import math
```

```python
train_data = pd.read_csv(r"C:\Users\miriy\Downloads\sonar_train.data", header = None)
test_data = pd.read_csv(r"C:\Users\miriy\Downloads\sonar_test.data", header= None)
valid_data = pd.read_csv(r"C:\Users\miriy\Downloads\sonar_valid.data", header = None)
```

```python
train_data[60].replace(2,-1, inplace = True)
test_data[60].replace(2,-1, inplace = True)
valid_data[60].replace(2,-1, inplace = True)
```

```python
train_data = np.array(train_data)
test_data = np.array(test_data)
valid_data = np.array(valid_data)
```

```python
def gaussian(data,mu,sigma):
    e = np.exp(-((data[0:60]-mu)**2)/(2*(sigma**2)))
    return np.log((1/(math.sqrt(2*math.pi)*sigma))*e)
```

```python
def accuracy(data,pred):
    return sum(data.T[60]*pred>0)/len(data)*100
```

```python
def stats(data):
    cls, counts = np.unique(data.T[-1],return_counts=True)
    l =len(data)
    prop =counts/l
    cls_prop = dict(zip(cls,prop))
    mean={}
    std={}
    for i in cls:
        mean[i] = np.mean(data[data[:,60]==i][:,0:60].T,axis=1)
        std[i] = np.std(data[data[:,60]==i][:,0:60].T,axis =1)
    return cls,cls_prop,mean,std
```

```python
cls,cls_prop,mean,std = stats(train_data)
```

```python
pred=[]
for j in test_data:
    prob_1= math.log(cls_prop[-1])+sum(gaussian(j,mean[-1],std[-1]))
    prob_2 = math.log(cls_prop[1])+sum(gaussian(j,mean[1],std[1]))
    if prob_1>prob_2:
        pred.append(-1)
    else:
        pred.append(1)
```

```python
acc = accuracy(test_data,pred)
```
